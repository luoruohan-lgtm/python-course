<img src='./Images/2_1.png' width=400  style='display: block; margin: 0 auto;'>

安装好 Python 后，可以通过两种方式编写和运行 Python 程序：

* **实验和测试用的交互式命令行模式**

交互式命令行模式下，通常会有提示符“>>>”，通常需要将代码输入到提示符后面，例如：'金融' + '科技'，然后按下回车键确认并运行代码：
```python
>>> '金融' + '科技'
```
上面这段代码中，提示符“>>>”是不需要输入的。第二行是程序运行的结果，即在屏幕上打印出字符串“金融科技”。这是因为在Python中，可以使用加号“+”将两个字符串连接后构成新的字符串“金融科技”。在交互式命令行模式下，会将程序运行结果直接反馈（打印到屏幕上）。

* **可多次运行的源文件模式**

当程序代码较多或者希望将程序保存下来多次运行时，通常需要使用文件来存储代码。包含Python代码的文件称为**源文件**，一般来说是以“.py”为后缀的文本文件。

你可以打开任意的文本文件编辑器，新建文件并命名为hello.py，保存在你的工作目录下，并在文件中输入以下Python代码：
```python
import sys
print(sys.version)
print('你好，Python')
```
在操作系统的终端中，切换到你的工作目录下，使用python命令可以运行这个程序。

### 1.4.2 使用IDLE开发

<a href="https://www.xuetangx.com/learn/swufe08091006588/swufe08091006588/26288051/video/66581726" style="color:#BDC3C7; text-decoration:none; font-size: 10pt;">《观看在线视频》</a>

IDLE被集成到标准Python当中，是一个能够编辑、运行和调试Python程序的用户图形界面。可以在安装了标准Python的计算机中找到启动IDLE的菜单项（Windows系统的开始菜单，或Mac OS X的启动台）。也可以在命令行或终端窗口中输入命令：idle 来启动。

<img src='./Images/2_2.png' width=500  style='display: block; margin: 0 auto;'>

使用IDLE等专业的集成工具进行程序开发时，会使用语法相关的颜色高亮。对于程序中不同类型的内容使用不同颜色，以便于开发者进行区分。例如，在默认情况下使用桔色显示关键字、红色显示注释语句和紫色显示内置函数等。

**IDLE还有以下功能特性：**
- 在Windows系统下使用<Alt-P>和<Alt-N>组合键查找历史命令（在Mac OS X系统中对应的是<Ctrl-P>和<Ctrl-N>组合键）
- 将光标放到窗口历史命令上，按下回车键<Enter>，可以复制该命令行
- 在输入时，按下<Tab>键，单词会自动补全
- 在使用函数时，输入函数名后的“(”后，会弹出关于该函数参数的提示信息

**IDLE常用快捷键**
在程序开发过程中，合理使用快捷键不但可以减少代码的错误率，而且可以提高开发效率。在 IDLE 中，可通过选择“Options -> Configure IDLE”菜单项，在打开的“Settings”对话框的“Keys”选项卡中查看。下表列出了 IDLE 中一些常用的快捷键：

|快捷键|说明|

|Alt+N|	浏览历史命令（下一条）|
|Alt+/|	自动补全前面曾经出现过的单词，如果之前有多个单词具有相同前缀，可以连续按下该快捷键，在多个单词中间循环选择|
|Alt+3|	注释代码块|

|Ctrl+Shift+Z|	恢复上—次的撤销操作|
|Ctrl+S|	保存文件|
|Ctrl+]|	缩进代码块|
|Ctrl+\[|	取消代码块缩进|
|Ctrl+F6|	重新启动 Python Shell|

**两种常用的单元格类型**

>* code：在单元格前面有In\[ \]的标识，通常用于输入和运行代码。
>* markdown：可以输入markdown语法，普通文本，通常用于输入描述性问题或者图表等内容。


**单元格的两种状态（键盘输入模式）**

>* 编辑模式：允许您将代码或文本输入到一个单元格中，并通过一个绿色的单元格来表示。单元格在编辑状态时，可以输入代码（code单元格）或者文本、图表（markdown单元格），输入完成后，通过菜单、Ctrl+Enter或Shift+Enter运行。
>* 命令模式：命令模式将键盘与命令绑定在一起，并通过一个灰色的单元格边界显示，该边框为蓝色的左边框。在命令模式下，可以通过键盘的一些快捷键添加单元格、删除单元格等。


**两种状态的切换**

>* 在编辑模式下，按ESC键切换至命令模式。
>* 在命令模式下，按Enter键切换至编辑模式。

**命令模式下，常用命令：**

>- 按下`A键`，在当前单元格上面插入代码块
>- 按下`B键`，在当前单元格下面插入代码块
>- 按下`X键`或者连续按两次`D键`，删除选中单元格
>- 按下`Y键`，将单元格变为代码类型
>- 按下`M键`，将单元格变为MarkDown类型
>- 按下`L键`，显示或隐藏单元格的行号

更多命令，请查看帮助菜单中的`快捷键`。

**使用MarkDown**


它是一种可以使用普通文本编辑器编写的标记语言，通过简单的标记语法，它可以使普通文本内容具有一定的格式。由于它的功能比纯文本更加强大，因此也有很多人用它来写博客。
在Jupyter的文本编辑单元格中，采用的就是Markdown的语法规范，因此它可以设置文本格式、插入链接、图片甚至数学公式。

创建一个单元格后，通过“单元格”菜单，将单元格类型更改为markdown，更改成功后，单元格开头没有“In[?]:”的提示符。然后点击单元格，按照markdown语法来输入文本。

常用的MarkDown语法：
 
>- 一个“#“表示一级标题，两个“#”表示二级标题，三个“#”表示三级标题，以此类推
>- 用两个空行来换行
>- 用反向逗号\`(esc按键)输入代码或符号
>- 以``![图片标签](图片源：本地文件或者链接)``格式来添加图片，以 ``<img src>`` 语句插入本地图片，可以设置高度与宽度 ``<img src="./test.jpg", width=100, heigth=100>``
>- 用\* 或+或- 加空格来表示无序列表
>- 用1.表示有序列表，即使数字不是按顺序的，运行后也会变为按顺序的
>>1. 列表项
>>1. 列表项
>- 输入\*\*\*\*\*或者\-\-\-\-\-表示分割线
>> ****
>- ``*斜体部分*``来达到*斜体效果*
>- ``**黑体加粗部分**``来达到**黑体加粗效果**
>- 插入一段代码需要三个反向逗号\`包裹，可以指定代码类型
>> ```Python
>> print('Hello, Python')
>> ```
>- 可以插入表格，表格的语句上一行必须为空行，不然表格不生效。
>> ``| 标题1 | 标题2 | 标题3 | |------| ------|------| | 短文本 | 中等文本 | 稍微长一点的文本 | | 稍微长一点的文本 | 短文本 | 中等文本 |``

> | 标题1 | 标题2 | 标题3 |
