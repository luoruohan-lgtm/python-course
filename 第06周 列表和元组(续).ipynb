
除了两个方面，“元组”数据类型几乎与列表数据类型一样。
#### 1）用圆括号( )表示
首先，元组输入时用圆括号``()``，而不是用方括号``[ ]``

````python
tuple('浦发银行')
````

#### 2）不可变性
元组与列表的主要区别还在于，元组像字符串一样，是***不可变的***。元组不能让它们的值被修改、添加或删除

````python
bank_name = tuple(['浦发银行', '招商银行', '保利地产', '平安银行'])
bank_name[1] = '农业银行'
````

如果元组中只有一个值，你可以在括号内该值的后面跟上一个逗号，表明这种情况。否则，Python 将认为，你只是在一个普通括号内输入了一个值。逗号告诉
Python，这是一个元组（不像其他编程语言，Python 接受列表或元组中最后表项后面跟的逗号）。

````python
t = [1]
print(type(t))
print(t)
````

````python
myStock = ('招商银行')
type(myStock)
````

````python
myStock = ('招商银行',)
type(myStock)
````

利用你可以用元组告诉所有读代码的人，你不打算改变这个序列的值。如果需要一个永远不会改变的值的序列，就使用元组。使用元组而不是列表的第二个好处在于，因为它们是不可变的，它们的内容不会变化，Python 可以实现一些**优化**，让使用元组的代码比使用列表的代码更快。

### 3.4.1　创建元组
在 Python 中，创建列表通常用 ``tuple`` 函数或圆括号 ``()``。
#### 1. tuple 函数
与 ``list`` 函数类似，``tuple`` 可以将参数中的可迭代对象转换成元组：

````python
tuple('浦发银行')
````

````python
type(('浦发银行',))
````

````python
tuple(['浦发银行', '招商银行', '保利地产', '平安银行'])
````

#### 2. 圆括号
使用圆括号 ``()`` 创建列表时，需要在列表中列出所有元素：

````python
('浦发银行', '招商银行', '保利地产', '平安银行')
````

直接使用圆括号 ``()`` 将得到一个空的元组：

````python
()
````

### 3.4.3　元组特性
在列表操作中， 除了对列表进行原地修改的方法， 其他方法都可以应用在元组中， 例如索引、分片、连接和重复等，与列表不同的是，在元组中使用这些方法，会返回新的元组，而不是列表。

#### 1. 单个元素的元组
如果仅仅使用圆括号把一个值括起来， Python 会认为是表达式， 得到的不会是元组类型，而是值本身的类型：

````python
x = (10.8)
type(x)
````

所以，如果确实希望得到一个包含单个元素的元组，需要在这一单个元素后面加上一个逗号，与表达式区分开来：

````python
x = (10.8,)
x
````

````python
type(x)
````

#### 2. 可以省略的圆括号
在不引起二义性的情况下，可以省略圆括号：

````python
stock_info = '600000', '浦发银行', 11.52, 11.54, 11.61, 11.4
stock_info
````

````python
type(stock_info)
````

#### 3. 不可变性
不可变性也是元组和列表最大的区别。 如果试图修改元组的某个元素， 将会触发 TypeError 的异常：

````python
bank_name = tuple(['浦发银行', '招商银行', '保利地产', '平安银行'])
bank_name[1] = '农业银行'
````

元组的不可变性仅指的是不可以改变元组的顶层元素。如果元组中的元素本身是可变的，Python 允许修改其内容：

````python
rate_info = (['活期存款',0.0030],['整存整取一年',0.0195],
             ['整存整取二年',0.0240],['整存整取三年',0.0280])
rate_info[1][1] = 0.0200
rate_info
````

### 3.4.4　为什么需要元组
主要原因如下：
* 元组占用的空间较小
* 你不会意外修改元组的值
* 可以将元组用作字典的键
* 函数的参数是以元组形式传递的

#### <font color=#A93226>【课堂案例】</font> 销售数据统计

<font size=3 face="楷体-简">统计Data/Sales目录下的文件2020_02.csv的销售数据，打印出销售金额最高的10个订单列表，含订单号和销售金额。</font><br><br>

<font size=1 color="gray">
注：在这个案例中，将提前了解一些关于循环和文件的内容，不能掌握也没有关系，后面章节会详细讲解。</font>

````python
import os
saleStat = [] # 列表结构：[[订单号, 销售额],...]
saleFile = open('Data/Sales/2020_02.csv', encoding='utf8')
saleFile.readline()
for line in saleFile:
    detail = line.replace('"', '').strip('\n').split(',')
    orderDetail = []
    for item in saleStat:
        if item[0] == detail[0]:
            orderDetail = item
            break
    if len(orderDetail) == 0:
        saleStat.append([detail[0], float(detail[3])*(int(detail[-2])-int(detail[-1]))])
    else:
        orderDetail[-1] += float(detail[3])*(int(detail[-2])-int(detail[-1]))
saleStat[:10]
````

````python
for item in saleStat:
    item.reverse()  # 把销售金额放在前面，便于排序
saleStat.sort(reverse=True)
````

````python
print("销售金额前10名的订单是：")
for item in saleStat[:10]:
    print("{:<10}{:>.2f}元".format(item[1], item[0]))
````

<font color=#145A32><b>【体验Pythonic】</b></font>  <font size=3 face="楷体-简">

<font size=3 color="gray">学完控制结构、函数和字典后，这个问题更加容易，也能统计更多的销售文件。例如，统计Data/Sales目录下所有销售数据，打印出销售金额最高的10个订单列表，含订单号和销售金额：</font>


````python
import os
fileNameList = [item for item in os.listdir('Data/Sales/') 
                if item.endswith('.csv') and not item.startswith('.')]
saleStatDic = {} # 字典结构：{订单号:销售额,...}
for fileName in fileNameList:
    saleFile = open('Data/Sales//' + fileName, encoding='utf8')
    saleFile.readline()
    for line in saleFile:
        detail = line.replace('"', '').strip('\n').split(',')
        saleStatDic[detail[0]] = saleStatDic.get(detail[0], 0)\
                                 + float(detail[3])*(int(detail[-2])-int(detail[-1]))        
saleStatList = list(saleStatDic.items())
saleStatList.sort(key=lambda item:item[1], reverse=True)
print("销售金额前10名的订单是：")
for item in saleStatList[:10]:
    print("{:<10}{:>.2f}元".format(item[0], item[1]))
````

## 【引导案例解析】
根据对引导案例的分析， 每个月学习用品支出最多的三项会有所变化。 最后， 打印出的是指定种类的金额和占比， 利用两个可变的列表按顺序分别存储名称和金额， 这样便于利用名称的位置得到对应的金额。

程序代码如下：

````python
name_list = []
amount_list = []
print('请输入支出最多的三项学习用品，名称和金额用逗号分割：')
tmp_str = input('第一项学习用品支出：')
name, amount = tmp_str.split(',')
name_list.append(name)
amount_list.append(float(amount))
tmp_str = input('第二项学习用品支出：')
name, amount = tmp_str.split(',')
name_list.append(name)
amount_list.append(float(amount))
tmp_str = input('第三项学习用品支出：')
name, amount = tmp_str.split(',')
name_list.append(name)
amount_list.append(float(amount))
search_name = input('请输入需要查询的学习用品种类名称：')
search_idx = name_list.index(search_name)
total_amount = sum(amount_list)
search_amount = amount_list[search_idx]
print('-' * 40)
print('学习用品中{}支出了{:.2f}元，占比：{:.2%}'.format(search_name,
                                          search_amount, search_amount / total_amount))
````
