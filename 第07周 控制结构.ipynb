### 缩进的作用
在 Python 中， 要求使用缩进来标识代码块。 相同层级的代码必须以垂直对齐的方式来组织， 即相同的缩进。

一些代码行可以作为一组，放在“代码块”中。 <font color="red">**可以根据代码行的缩进，知道代码块的开始和结束。**</font>  代码块有3 条规则。<br>
(1) 缩进增加时，代码块开始。<br>
(2) 代码块可以包含其他代码块。<br>
(3) 缩进减少为零，或减少为外面包围代码块的缩进，代码块就结束了。<br><br>

Python语言采用严格的“缩进”来表明程序的格式框架。缩进指每一行代码开始前的空白区域，用来表示代码之间的包含和层次关系。

Python 对于缩进的数量没有严格规定，比较常见的是使用 4 个空格或者 1 个制表符。但最好不要在同一段代码中混合使用空格和制表符，这样可能会引起语法错误。

在 Python 中，复合语句的首行通常以冒号“ **:** ”结尾。输入冒号后换行，编辑器会自动缩进后续行，用于定义代码块中的语句，从而清晰地组织代码结构。

```python
statement:
    code_block
```

# 4.2 条件表达式

控制语句的开始部分通常是“条件表达式”。分支语句通过条件表达式确定选择执行的代码块；循环语句通过条件表达式决定是否执行循环体中的代码块。通常来说，条件表达式的结果为一个布尔值，即True或者False，例如关系运算或者逻辑运算的表达式。控制语句根据条件是True还是False来决定做什么。

当表达式的最终值是下表中的某个值时，会被认作是假值：

|对象类型|值| &nbsp; |对象类型|值|
|:--|:--|----|:--|:--|
|**布尔**|False|&nbsp;|**null类型**|None|
|**整型**|0|&nbsp;|**浮点型**|0.0|
|**空字符串**|''|&nbsp;|**空列表**|[]|
|**空元组**|()|&nbsp;|**空字典**|{}|
|**空集合**|set()|&nbsp;|||

除了上表所列出来的值，其他的值都会被认作是真值。

## 4.3 分支结构

分支结构是程序根据条件判断结果而选择不同向前执行路径的一种运行方式，包括单分支结构和二分支结构。由二分支结构会组合形成多分支结构。

### 4.3.1  单分支结构：if语句
单分支结构的条件为真值True时，执行代码块，然后执行后续代码；而条件为假值False时，跳过代码块，继续执行后续代码。

单分支结构的语法如下所示：
```python
if condition:
    statements
```

<font color=#A93226><b>【例 4-1】</b></font> <font size=3 face="楷体-简">商品价格不应该为负数。 因此， 在用户输入商品价格的时候， 如果出现负数，则自动更正为 0。</font>

程序代码如下所示：

````python
price = float(input('请输入商品价格：'))
if price < 0:
    price = 0
    print('输入价格有误，已更正为0元')
print('商品价格是：{:.2f}元'.format(price))
````

### 4.3.2  else 语句<br>
if 子句后面有时候也可以跟着else 语句。只有if 语句的条件为False 时，else子句才会执行。在英语中，else 语句读起来可能是：“如果条件为真，执行这段代码。否则，执行那段代码”。else 语句不包含条件，在代码中，else 语句中包含下面部分：
```python
if condition:
    statements1
else:
    statements2
```

<font color=#A93226><b>【例 4-2】</b></font> <font size=3 face="楷体-简">身份号码中含有丰富的信息， 其中倒数第 2 位代表性别。 倒数第 2 位为奇数代表男性，偶数代表女性。编写代码，根据用户输入的身份证号码判定性别。</font>

程序代码如下所示：

````python
idCard = input('请输入身份证号码：')   # 520125197907167561
genderFlag = int(idCard[-2])
if genderFlag % 2:
    print("男")
else:
    print("女")
````

### 4.3.3  多分支结构：elif子句<br>
虽然只有if 或else 子句会被执行，但有时候可能你希望，许多可能的子句中有一个被执行。elif 语句是“否则如果”，总是跟在if 或另一条elif 语句后面。它提供了另一个条件，仅在前面的条件为False 时才检查该条件。在代码中，elif 语句总是包含以下部分：
```python
if condition1:
    statements1
elif condition2:
    statements2
else:
    statements3
```

<font color=#A93226><b>【例 4-3】</b></font> <font size=3 face="楷体-简">输入身份证号码后，为了确保输入的是正确的 18 位号码， 需要首先对身份证号码的长度进行判断。</font>

程序代码如下所示：

````python
idCard = input('请输入身份证号码：')
if len(idCard) != 18:
    print("身份证号码长度不是18位。")
elif int(idCard[16]) % 2:
    print("男")
else:
    print("女")
````

### 4.3.4 三元表达式: if/else
Python中的三元表达式可以将if-else语句放到一行里。语法如下：
```Python
True_expression if condition else False_expression
```
真值表达式或假值表达式可以是任何Python代码。它和下面的代码效果相同：
```Python
if condition:
    a = True_expression 
else:
    a = False_expression
```

利用三元表达式，【例 4-2】可以改写成如下形式：

````python
idCard = input("请输入身份证号码：") 
genderFlag = int(idCard[-2]) 
gender = "男" if genderFlag % 2 else "女"
print(gender)
````

## 4.3.5 <font color=#A93226>【案例】</font>汇率换算

汇率即外汇利率，指的是两种货币之间兑换的比率，也可以看作是一个国家的货币对 另一种货币的价值。具体是指某国货币与另外一个国家货币的比率或比价，或者说是用某 国货币表示的另外一个国家货币的价格。汇率会因利率、通货膨胀、国家的政治和每个国 家的经济等而变动。汇率是由外汇市场决定的。

汇率变动对国家进出口贸易有着直接的调节作用。在一定条件下，通过使本国货币对 外贬值， 即让汇率下降， 会起到促进出口、限制进口的作用；反之， 本国货币对外升值， 即汇率上升，则起到限制出口、增加进口的作用。

<font size=3 face="楷体-简">编写程序，根据用户的输入，计算人民币和美元之间的换算金额，实现以下功能：
- <font size=3 face="楷体-简">用户输入 ¥ 开头的人民币金额时，输出换算后的美元金额。
- <font size=3 face="楷体-简">用户输入 $ 开头的美元金额时，输出换算后的人民币金额。
- <font size=3 face="楷体-简">用户输入的内容以其他字符开头，打印出错误信息。

<font size=3 face="楷体-简">假定 100 美元换算人民币金额为 692.3 元，即美元兑人民币的汇率为 6.923。
</font>

程序代码如下所示：

````python
exchange_rate = 6.923
amount_str = input('请输入兑换金额(美元以$开头，人民币以¥开头):')
if amount_str[0] == '$':
    amount = float(amount_str[1:]) * exchange_rate
    print('{}可以兑换人民币{:.2f}元'.format(amount_str,amount))
elif amount_str[0] == '¥':
    amount = float(amount_str[1:]) / exchange_rate
    print('{}可以兑换{:.2f}美元'.format(amount_str,amount))
else:
    print('输入格式错误。')
````

## 4.4 while 循环语句
<a href="https://www.xuetangx.com/learn/space/swufe08091006588/swufe08091006588/29598038/video/75642699" style="color:#BDC3C7; text-decoration:none; font-size: 10pt;">《观看在线视频》</a>



Python 中有 while 和 for 两种循环结构，利用循环语句，可以让一个代码块一遍又一遍的执行。

对于while循环，**只要while语句的条件为True，while子句中的代码就会执行**。Python当中，while实现循环的方法如下：<br>
```Python
while condition:
    statements
```
while 语句看起来和if 语句类似。不同之处是它们的行为。if 子句结束时，程序继续执行if 语句之后的语句。但在while 子句结束时，程序执行跳回到while 语句开始处。while 子句常被称为“while 循环”，或就是“循环”。例如下面的代码：

````python
userCounter = 0
if userCounter < 10:
    userCounter = userCounter + 1
    print("第{}个用户".format(userCounter))
````

````python
userCounter = 0
while userCounter < 10:
    userCounter += 1
    print("第{}个用户".format(userCounter))
````

<font color=#A93226><b>【例 4-4】</b></font> <font size=3 face="楷体-简">在超市购物结算时， 收银员逐一扫描商品上的条码， 记录购买的商品和数量，最终累加出总的金额。每个人购买的商品数量不一样，在开始扫码时，收银员也不清 楚有多少个商品。 因此扫码记录不断地重复， 直到按下确认按钮。 </font>

<font size=1 color="gray">注：这就是一个循环的过程，在这种不太确定循环次数的场景中，最好使用 while 循环。</font>

程序代码如下所示：

````python
total_amt = 0
goods_str = input('请输入商品数量和单价（用逗号隔开，直接回车结束）：')
while len(goods_str) > 0:
    goods_info_list = goods_str.split(',')
    total_amt += int(goods_info_list[0]) * float(goods_info_list[1])
    goods_str = input('请输入商品数量和单价（用逗号隔开，直接回车结束）：')
print('需支付金额：{:.2f} 元'.format(total_amt))
````

## 4.5 for 循环语句
<a href="https://www.xuetangx.com/learn/space/swufe08091006588/swufe08091006588/29598038/video/75642700" style="color:#BDC3C7; text-decoration:none; font-size: 10pt;">《观看在线视频》</a>



通常，for 循环用于遍历序列（字符 串、列表和元组等）或任何可迭代对象内的元素，是一个通用的迭代器。for循环的一般形式是：
```python

```

<font color=#A93226><b>【例 4-5】</b></font> <font size=3 face="楷体-简">几何平均收益率是将各单个期间的收益率相乘， 然后开 n 次方。 几何平均收益率使用了复利的思想，即考虑了资金的时间价值。这个平均收益指标优于算术平均收益率，因为它引入了复利，即通过对时间进行加权来衡量最初投资价值的复合增值率，从而克服了算术平均收益率有时会出现的上偏倾向。几何平均收益率的计算公式为：

$$
\overline{R} = \sqrt[n]{\prod_{i=1}^{n}{(1+R_{i})}} - 1
$$

<font size=3 face="楷体-简">其中， $\overline{R}$ 是几何平均收益率， $R_i$ 是每期收益率。 

<font size=3 face="楷体-简">已知招商银行在 2020 年 7 月 20 日至 2020 年 7 月 24 日的每日涨跌幅（收益率）分别 为：1.19%、–0.43%、–0.40%、–0.84% 和 –1.00%。 编写程序， 计算招商银行这几天的几 何平均收益率。
</font>

程序代码如下所示：

````python
return_list = [0.0119, -0.0043, -0.004, -0.0084, -0.01]
total = 1.0
for r in return_list:
    total = total * (1 + r)
    return_avg = total ** (1 / len(return_list)) - 1
print('招商银行几何平均收益率为：{:.2%}'.format(return_avg))
````

有一种遍历方式是利用序号（偏移量），此时需要将包含列表序号的可迭代对象放在for循环首行的in关键字后面，通常利用range函数来实现。
range函数会得到一个按需产生整数元素的可迭代对象，其语法结构有两种形式：

range(stop)

````python
return_list = [0.0119, -0.0043, -0.004, -0.0084, -0.01]
total = 1.0
for i in range(0, 5):
    total = total * (1 + return_list[i])
return_avg = total ** (1 / len(return_list)) - 1
print('招商银行几何平均收益率为：{:.2%}'.format(return_avg))
````

## break语句
break用来跳出最内层for或while循环，脱离该循环后程序从循环后代码继续续执行。

````python
#nameStr = "swufe"
nameStr = "abcde"
for c in nameStr:
    if c == 'u':
        print("在{}中找到u了。".format(nameStr))
        break
else:
    print("{}中没有u字母。".format(nameStr))
````

每个break语句只有能力跳出当前层次循环。

````python
for c in "swufe":
    for i in range(10):
        print(c, end="")
        if c=='u':
            break
````

## continue 语句
像break 语句一样，continue 语句用于循环内部。如果程序执行遇到continue语句，就会马上 **跳回到循环开始处，重新对循环条件求值** （这也是执行到达循环末尾时发生的事情）。

试试下面的用户验证：

````python
while True:
    print('请输入用户名：')
    name = input()
    if name != 'zhangsan':
        continue
    print('Hello, zhangsan. 密码是什么? (学校简称.)')
    password = input()
    if password == 'swufe':
        break
print('用户验证成功.')
````

<font color=#A93226><b>【例 4-6】</b></font> <font size=3 face="楷体-简">股票代码和名称按照相同的顺序分别存储在两个元组中。 编写程序，根据输入的股票代码找出股票名称。如果直接回车（输入空字符串），则结束程序。
</font>

程序代码如下所示：

````python
stock_code = ('600000', '600036', '600048', '000001')
stock_name = ('浦发银行', '招商银行', '保利地产', '平安银行')
while True:
    search_code = input('请输入股票代码：')
    if len(search_code) == 0:
        break
    if search_code not in stock_code:
        print('代码没有在股票池中。')
        continue
    print('{}的股票名称是：{}'.format(search_code,stock_name[stock_code.index(search_code)]))
````

## else 语句
在有些场景中，需要根据循环结束的原因（正常结束还是因为执行break而结束）来作出不同的操作。在其它程序设计语言中，通常使用一个标记变量来实现。在Python的循环，加入了else子句，使得程序编写更加方便：

<font color=#A93226><b>【例 4-7】</b></font> <font size=3 face="楷体-简">素数， 又称为质数， 指的是除了 1 和它本身外， 不能被任何整数整除的数， 例如 5 就是素数。编写程序，找出 50 以内的素数。
</font>

程序代码如下所示：

````python
result = [] 
for i in range(2,51):
    for j in range(2, i):
        if i % j == 0:
            break
    else:
        result.append(i) 
print(result)
````

### 4.7.2　列表推导式
推导式是从可迭代对象中快速简洁地创建数据类型的一种方法。它使得你用优美简短的代码就能实现循环甚至条件判断。

利用推导式可以得到列表、字典、集合以及生成器。列表推导式会产生一个新的列表，其语法形式如下所示：
```python
[expr for item in iterable_obj]
```


<font color=#A93226><b>【例 4-8】</b></font> <font size=3 face="楷体-简">用户输入或者从文件读取的数据通常都是字符串类型， 如果需要进行数学运算，需要将字符串转换为数字类型后再运算。编写程序，将输入的销售数量进行累加。
</font>

程序代码如下所示：

````python
amt_str = input('请输入销售数量(用逗号隔开):') 
amt_str_list = amt_str.split(',') 
amt_int_list = [int(amt) for amt in amt_str_list] 
print('总销售数量为:{}件'.format(sum(amt_int_list)))
````

列表推导式也可以加上 if 关键字， 对原可迭代对象中的元素进行判断， 符合条件的元素在使用 expr 表达式运算后加入新列表中：
```python
[expr for item in iterable_obj if condition]
```

对于例 4-8 来说，为了防止输入的销售数量是负数，可以对代码进行以下改造：

````python
amt_str = input('请输入销售数量(用逗号隔开):') 
amt_str_list = amt_str.split(',') 
amt_int_list = [int(amt) for amt in amt_str_list if '-' not in amt] 
print('总销售数量为:{}件 '.format(sum(amt_int_list)))
````

### 4.7.3　并列遍历：zip 函数
Python 中提供了一个非常有用的函数：zip 函数。 这个函数可以将这些序列并排的元素配对成元组后，组成一个新的可迭代对象。其语法格式如下：
```python
zip(*iterables)
```


````python
stock_code = ('600000', '600036', '600048', '000001') 
stock_name = ('浦发银行', '招商银行', '保利地产', '平安银行') 
close_price = [11.54, 36.84, 16.61, 14.49] 
for item in zip(stock_code, stock_name, close_price):
    print('{1} 的代码是：{0}，收盘价：{2:.2f} 元'.format(item[0], item[1], item[2]))
````

### 4.7.4　简单循环的替身：map 函数
在 Python 中，可以利用 map 函数简化这类代码。其语法格式如下：
```python
map(func, *iterables)
```
map 函数有两个参数，func 是一个函数， 也就是对序列中元素进行的操作。 例如， 例 4-8 的代码可以进行如下改写：

````python
amt_str = input(' 请输入销售数量 ( 用逗号隔开 )：') 
amt_str_list = amt_str.split(',') 
amt_int_map_obj = map(int, amt_str_list) 
print(' 总销售数量为：{} 件 '.format(sum(amt_int_map_obj)))
````

## 【引导案例解析】
根据对引导案例的分析， 结合本章的学习， 利用 while 循环结构和 if 分支结构对 第 3 章引导案例改进，使得程序支持任意项明细情况的输入。

程序代码如下：

````python
name_list = []
amount_list = []
print('请输入学习用品支出情况，名称和金额用逗号分割，直接回车结束输入。')
print('-' * 40)
i = 1
while True:
    tmp_str = input('第{}项：'.format(i))
    if len(tmp_str) == 0:
        break
    name, amount = tmp_str.split(',')
    name_list.append(name)
    amount_list.append(float(amount))
    i += 1
print('-' * 40)
search_name = input('请输入需要查询的学习用品种类名称：')
search_idx = name_list.index(search_name)
total_amount = sum(amount_list)
search_amount = amount_list[search_idx]
print('-' * 40)
print('学习用品中{}支出了{:.2f}元，占比：{:.2%}'.format(search_name,
                                          search_amount, search_amount / total_amount))
````
