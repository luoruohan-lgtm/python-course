# <div align="center">关于本课程的说明</div>

## 1. 计算机公共基础课程：
* 计算机基础（解决计算机操作的问题）
* 程序设计及应用（解决计算机应用的问题）

## 2. 课程考核
### 期末考试(50%)：
* **考核内容：** 教学内容中需要理解和掌握的内容；考察学生的自主学习能力；考察学生的程序设计应用能力。
* **考核方式：** <font color=red>由课程组统一组织</font>。
* **占比分值：** 50分


### 在线成绩(20%)：

* **考核内容：** 学堂在线部分视频（课件要求的视频）、章节作业、在线期末考试、论坛活跃度
* **考核方式：** <font color=red>由平台提供，任课老师自主决定比例</font>。
* **占比分值：** 20分

### 平时成绩(30%)：

|考试形式|说明|考察方式|分值|
|----|:--|----|--|
|教学过程|考勤、讨论、回答问题、期中考试、课堂测验等|方式由任课老师自主决定|10分|
|课后作业|全学期不少于5次|方式由任课老师自主决定|20分|
